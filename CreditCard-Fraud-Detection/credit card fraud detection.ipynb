```python
# Import libraries 
import pandas as pd

#import display for DataFrame usage
from IPython.display import display

# Load the credit card transactions dataset
data = pd.read_csv("creditcard.csv")

# On Success - Display the first ten records
display(data.head(n=10))
```

```python
# Count total number of transactions 
print "Total number of records: {}".format(data.shape[0])
```

```python
display(data.describe())

#storing original data
total_data = data

#preserving only dimensions
data.drop(['Time', 'Amount', 'Class'], axis = 1, inplace = True)

display(data.describe())
```

```python
#import PCA
from sklearn.decomposition import PCA

#import visualizations from visuals class
import visuals as vs


display(data.head(n=10))


# Implementing Dimensionality Reduction
# Lets apply PCA by fitting the data with only 20 dimensions
pca = PCA(n_components=20)

pca.fit(data)
pca_data = pca.transform(data)

# Generate PCA results plot
pca_results = vs.pca_results(data, pca)
```

```python
#Calculating total explained variance
explained_variance = 0
for i in range(0,15):
    explained_variance += pca.explained_variance_ratio_[i]
print "Total explained variance is %f " % explained_variance
```
