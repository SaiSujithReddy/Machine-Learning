```python
#Download dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
df = pd.read_csv('../data.csv')
df.head(5)
#Checking types of data
print(df.dtypes)
#Dropping irrelevant columns
col_to_drop = ['Engine Fuel Type','Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size']
df = df.drop(col_to_drop, axis=1)
df.head(5)
#Renaming columns
to_be_renamed = {"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" }
df = df.rename(columns=to_be_renamed)
df.head(5)
#dropping duplicate rows
print(df.shape)
df = df.drop_duplicates()
print(df.shape)
#Find missing values
#print(df.count())
print(df.isnull().sum())
#Drppping missing values
df = df.dropna()
print(df.shape)
# Filling missing values

#Detecting outliers
sns.boxplot(x=df['Price'])

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
#Removing outliers
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
#Bar plots
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');

#Heat Maps
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
#Scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
